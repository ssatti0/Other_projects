## Training and Testing Data

Now that we've explored the data a bit, let's go ahead and split the data into training and testing sets.
** Set a variable X equal to the numerical features of the customers and a variable y equal to the "Yearly Amount Spent" column. **

```python
customers.columns
```

```python
feature_names = ['Avg. Session Length', 'Time on App',
       'Time on Website', 'Length of Membership']
```

```python
X = customers[feature_names]
y = customers['Yearly Amount Spent']
```

```python
y.head(3) #
```

** Use model_selection.train_test_split from sklearn to split the data into training and testing sets. Set test_size=0.3 and random_state=101**

```python
from sklearn.model_selection import train_test_split
```

```python
# train, test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Training the Model

Now its time to train our model on our training data!

** Import LinearRegression from sklearn.linear_model **

```python
from sklearn.linear_model import LinearRegression
```

**Create an instance of a LinearRegression() model named lm.**

```python
lm = LinearRegression()
```

** Train/fit lm on the training data.**

```python
lm.fit(X_train, y_train)
```

```python
lm.intercept_
```

**Print out the coefficients of the model**

```python
lm.coef_
```

## Predicting Test Data
Now that we have fit our model, let's evaluate its performance by predicting off the test values!

** Use lm.predict() to predict off the X_test set of the data.**

```python
predictions = lm.predict(X_test)
```

** Create a scatterplot of the real test values versus the predicted values. **

```python
plt.scatter(y_test, predictions, alpha=0.4)
plt.xlabel('Actual Yearly Amount Spent')
plt.ylabel('Predicted Yearly Amount Spent')
plt.title('Actual vs Predicted Yearly Amount Spent')
```

## Evaluating the Model

Let's evaluate our model performance by calculating the residual sum of squares and the explained variance score (R^2).

** Calculate the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error. Refer to the lecture or to Wikipedia for the formulas**

```python
# plot Residuals
plt.scatter(y_test, y_test-predictions)
plt.xlabel('y_test')
plt.ylabel('residuals')
```

```python
# error calculations
from sklearn import metrics
print('Mean Absolute Error: ', metrics.mean_absolute_error(y_test, predictions))
print('Mean Squared Error: ', metrics.mean_squared_error(y_test, predictions))
print('Root Mean Squared Error: ', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

## Residuals

You should have gotten a very good model with a good fit. Let's quickly explore the residuals to make sure everything was okay with our data. 

**Plot a histogram of the residuals and make sure it looks normally distributed. Use either seaborn distplot, or just plt.hist().**

#### Below is a histogram of the residuals, showing a normal distribution.

```python
sns.distplot((y_test-predictions),bins=50) 
```

## Conclusion
We still want to figure out the answer to the original question, do we focus our efforst on mobile app or website development? Or maybe that doesn't even really matter, and Membership Time is what is really important.  Let's see if we can interpret the coefficients at all to get an idea.

** Recreate the dataframe below. **

```python
Coefficients = pd.DataFrame(lm.coef_, X.columns)
Coefficients.columns = ['Coefficient']
Coefficients
```

** How can you interpret these coefficients? **

*Keeping all other features fixed, the coefficients can be interpreted as follows:
- A unit increase in Avg session length results in a yearly amount spent increase of about 26 dollars.
- A unit increase in Time in App results in a yearly amount spent increase of 38.5 dollars.
- A unit increase on Time on Website results in a yearly amount spent of 0.19 dollars
- A unit increase in length of membership results in a yearly amount spent of 61.27 dollars. *

**Do you think the company should focus more on their mobile app or on their website?**

*From the analysis conducted, it seems that the most important factor affecting Yearly Amount Spent is the length of membership, with a coefficient of 61.2, the second most important factor is Time Spend on App. Time on Website makes the least amount of contribution to yearly amount spent by the customers.
Therefore, it is recommended that the company focus on keeping those clients with memberships, and spending more on improving the mobile App experience. This is the short term solution (the low hanging fruit, so to speak).
Longer term, it is a good idea to examine the website and try to understand the reasons for its poor performance.*
